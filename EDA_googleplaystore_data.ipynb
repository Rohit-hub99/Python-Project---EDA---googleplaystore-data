```python
df['Installs'] = df['Installs'].str.replace(',','').str.replace('+','')
```

```python
df['Installs']
```

```python
# 1. Handling missing values
df = df.dropna()
```

```python
df
```

```python
# 2. Cleaning 'Installs' column
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '')
df['Installs'] = df['Installs'].astype(int)
```

```python
# 3. Cleaning 'Price' column
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = df['Price'].astype(float)
```

```python
df['Price']
```

```python
# 4. Cleaning 'Size' column

def size_to_mb(size):
    if size.endswith('M'):
        return float(size.replace('M', ''))
    elif size.endswith('k'):
        return float(size.replace('k', '')) / 1024
    else:
        return np.nan

df['Size'] = df['Size'].replace('Varies with device', np.nan)
df['Size'] = df['Size'].dropna().apply(size_to_mb)
```

```python
df['Size']
```

```python
# 5. Correcting 'Reviews' column type
df['Reviews'] = df['Reviews'].astype(int)
```

```python
df['Reviews']
```

```python
# 6. Correcting 'Rating' column type
df['Rating'] = df['Rating'].astype(float)
```

```python
df['Rating']
```

```python
# Plot top 10 categories by number of apps


top_categories = df['Category'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='viridis')
plt.title('Top 10 Categories by Number of Apps')
plt.xlabel('Number of Apps')
plt.ylabel('Category')
plt.show()
```

```python
# Distribution of App Ratings (Histogram)

plt.figure(figsize=(10,6))
sns.histplot(df['Rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of App Prices (Histogram)

plt.figure(figsize=(10,6))
sns.histplot(df[df['Price'] > 0]['Price'], bins=20, kde=True, color='green')
plt.title('Distribution of App Prices (Paid Apps Only)')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of App Sizes (Histogram)

plt.figure(figsize=(10,6))
sns.histplot(df['Size'].dropna(), bins=20, kde=True, color='orange')
plt.title('Distribution of App Sizes')
plt.xlabel('Size (MB)')
plt.ylabel('Count')
plt.show()
```

```python
# Rating vs Reviews (Scatter Plot)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Reviews', y='Rating', alpha=0.5)
plt.title('Rating vs Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.xscale('log')
plt.show()
```

```python
# Rating vs Installs (Scatter Plot)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Installs', y='Rating', alpha=0.5, color='red')
plt.title('Rating vs Installs')
plt.xlabel('Number of Installs')
plt.ylabel('Rating')
plt.xscale('log')  # Installs also huge, use log scale
plt.show()
```

```python
# Price vs Installs (Scatter Plot)

plt.figure(figsize=(10,6))
sns.scatterplot(data=df[df['Price'] > 0], x='Price', y='Installs', alpha=0.5, color='green')
plt.title('Price vs Installs (Paid Apps Only)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Installs')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# Category vs Average Rating (Bar Plot)

category_rating = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,8))
sns.barplot(x=category_rating.values, y=category_rating.index, palette='coolwarm')
plt.title('Average Rating by Category')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```

```python
# Does app size affect rating?

df['Size_Category'] = pd.cut(df['Size'], bins=[0, 1, 10, float('inf')], labels=['Small', 'Medium', 'Large'])

size_rating = df.groupby('Size_Category')['Rating'].mean()
```

Conclusion


- The Google Play Store is dominated by apps in the FAMILY and GAME categories.
- Most apps are free, and free apps are rated similarly to paid apps.
- Highly popular apps belong to categories like COMMUNICATION and PRODUCTIVITY.
- There is little correlation between app size and rating.
- Most users prefer free apps, and very few paid apps achieve high install counts.

Overall, app type, category, and user engagement (reviews, installs) have a bigger impact on ratings than app size or price.
